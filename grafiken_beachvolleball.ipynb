```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Spielstrken der Mannschaften
a_strength = int(input("Bitte geben Sie eine Zahl zwischen 1-10 ein"))
b_strength = int(input("Geben Sie eine Zahl zwischen 1- 10 ein "))

# Anzahl der Spiele
num_games = 1000

# Matrix zur Speicherung der Ergebnisse
results_matrix = np.zeros((11, 11))

# Schleife ber alle mglichen Kombinationen von Spielstrke und Risikofaktor
for a_strength_idx in range(11):
    for a_risk_idx in range(11):
        a_strength = a_strength_idx
        a_risk = a_risk_idx
        a_wins = 0
        b_wins = 0
        for i in range(num_games):
            # Berechnung der Erfolgschancen der Mannschaften
            a_chance = a_risk/10 * a_strength/10 + (10-a_risk)/10 * (1-b_strength/10)
            b_chance = (10-a_risk)/10 * b_strength/10 + a_risk/10 * (1-a_strength/10)

            # Simulation des Ballwechsels
            if np.random.rand() < a_chance:
                a_wins += 1
            else:
                b_wins += 1

        # Speicherung der Ergebnisse
        results_matrix[a_strength_idx, a_risk_idx] = a_wins

# Erstellung der Heatmap
sns.heatmap(results_matrix, cmap="YlGnBu", annot=True, fmt=".0f")
plt.xlabel("Risikofaktor fr Mannschaft A")
plt.ylabel("Spielstrke von Mannschaft A")
plt.title("Anzahl der gewonnenen Spiele von Mannschaft A")
plt.show()
```
